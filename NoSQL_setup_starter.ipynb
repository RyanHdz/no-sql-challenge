5. Some of the number values are stored as strings, when they should be stored as numbers.

Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
# Change the data type from String to Decimal for longitude and latitude
from bson.decimal128 import Decimal128

# Find all documents in the 'establishments' collection
documents = establishments.find({})

# Loop through each document
for doc in documents:
    # Convert 'longitude' and 'latitude' to Decimal
    longitude = Decimal128(doc['geocode']['longitude'])
    latitude = Decimal128(doc['geocode']['latitude'])

    # Define the new values for 'longitude' and 'latitude'
    new_values = {"$set": {"geocode.longitude": longitude, "geocode.latitude": latitude}}

    # Update the document
    establishments.update_one({"_id": doc["_id"]}, new_values)

print("Updated all documents.")
```

Use `update_many` to convert `RatingValue` to integer numbers.

```python
# Set non 1-5 Rating Values to Null
documents = establishments.find({})

# Loop through each document
for doc in documents:
    # Check if 'RatingValue' exists and is not None and is between 1 and 5
    if 'RatingValue' in doc and doc['RatingValue'] is not None and 1 <= doc['RatingValue'] <= 5:
        # 'RatingValue' is already an integer, no need to convert
        rating_value = doc['RatingValue']
    else:
        # Set 'RatingValue' to None
        rating_value = None

    # Define the new value for 'RatingValue'
    new_values = {"$set": {"RatingValue": rating_value}}

    # Update the document
    establishments.update_one({"_id": doc["_id"]}, new_values)

print("Updated all documents.")
```

```python
# Change the data type from String to Integer for RatingValue
for doc in documents:
    # Check if 'RatingValue' exists and is not None
    if 'RatingValue' in doc and doc['RatingValue'] is not None:
        # Convert 'RatingValue' to an integer
        rating_value = int(doc['RatingValue'])
    else:
        # Keep 'RatingValue' as None
        rating_value = None

    # Define the new value for 'RatingValue'
    new_values = {"$set": {"RatingValue": rating_value}}

    # Update the document
    establishments.update_one({"_id": doc["_id"]}, new_values)

print("Updated all documents.")
```
